### Algorithms

The deep learning algorithms to be used in this project are as follows;

**1-) Convolutional Neural Network (CNNs)**

CNNs are fully connected feed forward neural networks.
CNNs are very effective in reducing the number of parameters without losing on the quality of models
Images have high dimensionality (as each pixel is considered as a feature) which suits the above described abilities of CNNs.
Also, it has dimensionality reduction feature.

**2-) Inception**

It is open source like Inception-v3.
They have been trained on millions of images with extremely high computing power,
which can be very expensive to produce from scratch.
This model is trained on an extremely large dataset,
so it can yield successful results with few train sets.so applying this model can save us time and performance
