## Dataset
Dataset Link: https://www.kaggle.com/datasets/jessicali9530/celeba-dataset

Dataset explanation was made by pulling data in the cells below. 
The data comes as a zip file. The zip file contains 4 csv files and photos in a folder. Using opendatasets library, dataset is downloaded from kaggle via api. Then, the files inside the folder were read. the following are the explanations according to the files read. This folder contains photos. Analysis will be done using photographs.

**img_align_celeba** : The folder where the photos are located. The process takes a long time because there are too many photos.

**list_attr_celeba.csv** : In this data, there are 40 different attributes related to photos. Labels change as 1 and -1. It is labeled as 1 if the relevant feature exists, -1 if it is not.

**list_bbox_celeba.csv** : Contains information about the dimensions of the photos.

**list_eval_partition.csv**: It has test-train and validation tags.

**list_landmarks_align_celeba.csv** : There are coordinates of the facial features of the people in the photos. For example, eyes, mouth, nose, etc.
